```python
import json
import matplotlib.pyplot as plt
import seaborn
plt.style.use('ggplot')
```

```python
file = open("fitness.txt", "r")
lines = file.readlines()
```

```python
fitness = list()
for i in range(len(lines)):
    l = lines[i][:-1]
    d = eval(l)
    fitness.append(d)
    
```

```python
values = list()
for i in range(len(fitness[0])):
    v = list()
    for j in range(len(fitness)):
        v.append(fitness[j][i])
    values.append(v)
```

```python
x = [i+1 for i in range(len(values))]
f = plt.figure(1, figsize=(15, 5))
plt.xlabel("Individuo")
plt.ylabel("Aptidao")
plt.title("Evolucao das aptidoes ao longo das geracoes")
for i in range(len(values)):
    plt.plot(x, [pt[i] for pt in values],label = 'id %s'%str(i+1))
    
plt.legend()
plt.show()
```
