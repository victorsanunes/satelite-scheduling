```python
import json
import matplotlib.pyplot as plt
import seaborn
plt.style.use('ggplot')
```

```python
file = open("fitness.txt", "r")
lines = file.readlines()

fitness = list()
for i in range(len(lines)):
    l = lines[i][:-1]
    d = eval(l)
    fitness.append(d)
    

values = list()
for i in range(len(fitness[0])):
    v = list()
    for j in range(len(fitness)):
        v.append(fitness[j][i])
    values.append(v)


individuo = 9
x = [i+1 for i in range(len(values[individuo]))]
f = plt.figure(num=None, figsize=(15, 6), dpi=160, facecolor='w', edgecolor='k')
# plt.scatter(x, values[individuo])
plt.plot(values[individuo])
# plt.figure(1, figsize=(35, 15))

# plt.grid(True)

plt.xlabel("Individuo")
plt.ylabel("Aptidao")
plt.title("Evolucao das aptidoes ao longo das geracoes")
f.savefig("fitness_25.png")
plt.show()
```

```python
def readValues(filename):
    file = open(filename, "r")
    lines = file.readlines()
    values = [float(x) for x in lines] 
    return values
```

```python
seed = 1
values_10 = readValues("values_10.txt")
values_15 = readValues("values_15.txt")
values_20 = readValues("values_20.txt")
values_25 = readValues("values_25.txt")
values_30 = readValues("values_30.txt")
```
