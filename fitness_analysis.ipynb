```python
def readValues(filename):
    file = open(filename, "r")
    lines = file.readlines()
    values = [float(x) for x in lines] 
    return values
```

```python
crossover_rate = "5"
populacao = "4"
base_path = "solicitacoes_com_prioridade/populacao"+populacao+"/crossover_"+crossover_rate

# values = readValues(base_path+"/best_fitness.csv")
values = pd.read_csv(base_path+"/best_fitness.csv", sep=',', header=None)
best = [2.425 for i in range(0, len(values))]
```

```python
seed = 1

fig, ax = plt.subplots()
x = [i+1 for i in range(len(values))]

line1, = ax.plot(x, best, '--', label="Valor timo")
line2, = ax.plot(x, values, label = "crossover="+crossover_rate+"%")
# line3, = ax.plot(x, values_15, label = "crossover=15%")
# line4, = ax.plot(x, values_20, label = "crossover=20%")
# line5, = ax.plot(x, values_25, label = "crossover=25%")
# line6, = ax.plot(x, values_30, label = "crossover=30%")
# line7, = ax.plot(x, ninetyPercent, '--', label="90% do valor timo")

ax.legend(loc='lower right')

plt.title("Aptido dos melhores indivduos em cada gerao(seed="+str(seed)+")")
plt.xlabel("Geraes")
plt.ylabel("Aptido")
fig.set_dpi(160)
fig.set_size_inches(16,6)
# fig.savefig("fitness_geral_seed"+str(seed)+"_com_prioridade.png")
fig.savefig(base_path+"/crossover("+ crossover_rate+")_seed("+str(seed)+")_prioridade(manual).png")
plt.show()
```
